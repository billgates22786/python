```python
!pip install -q dash plotly pandas
from dash import Dash, html, dcc, callback, Output, Input
import dash
from dash import dcc, html
from dash.dependencies import Input, Output
import plotly.express as px
import pandas as pd
df=pd.read_csv('Mart.csv')

app = Dash(__name__)
app.title = 'K-Mart'
app.layout = html.Div([
    html.H2("K-Mart Dashboard", style={'textAlign': 'center'}),

    dcc.Dropdown(
        id='Category-Dropdown',
        options=[{'label': c, 'value': c} for c in df['category'].unique()],
        placeholder="Please select a category",
        clearable=False,
        multi=True
    ),

    html.Div([
        dcc.Graph(id='sales-graph'),
        dcc.Graph(id='profit-pie'),
        dcc.Graph(id='discount-sunburst')
    ])
])


@app.callback(
    Output('sales-graph', 'figure'),
    Output('profit-pie', 'figure'),
    Output('discount-sunburst', 'figure'),
    Input('Category-Dropdown', 'value')
)
def update_charts(selected):
    filtered_df = df[df['category'].isin (selected)]

    fig = px.bar(
        filtered_df,
        x='subcategory',
        y='sales',
        title=f"Sales by Subcategory in '{selected}' Category",
        text='sales'
    )
    fig.update_traces(marker_color='red', textposition='outside')
    fig.update_layout(yaxis_title='Sales', xaxis_title='Subcategory')

    fig2 = px.pie(
        filtered_df,
        values='profit',
        names='subcategory',
        title=f"Profit by Subcategory in '{selected}' Category"
    )
    fig3 = px.sunburst(
        filtered_df,
        path=["category","subcategory"],
        values='discount',
        title=f"Discount Distribution in '{selected}' Category"
    )
    return fig, fig2,fig3



if __name__ == '__main__':
    app.run(debug=True)
```
