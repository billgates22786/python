```python
class Tolltax:
  def __init__(self,payingcars,nonpayingcars,collection):
    self.payingcars=payingcars
    self.nonpayingcars=nonpayingcars
    self.collection=collection
    self.payingcars=0
    self.nonpayingcars=0
    self.collection=0
    for i in range(1,12):
      print("Press 1 Paying Cars\nPress 2 Non Paying Cars\nPress 3 Collection")
      self.choice=int(input("Enter Your Choice: "))
      if self.choice==1:
        print("You have selected Paying Cars")
        self.payingcars+=1
        self.collection+=100
      elif self.choice==2:
        print("You have selected Non Paying Cars")
        self.nonpayingcars+=1
        self.collection-=100
      elif self.choice==3:
        print("You have selected Collection")
        print("Total Paying Cars:", self.payingcars)
        print("Total Non-Paying Cars:", self.nonpayingcars)
        print("TOTAL COLLECTION:", self.collection)
        break
      else:
        print("Invalid Option")


obj=Tolltax(0,0,0)

```
