```python
X = df.drop('price',axis='columns')
X
```

```python
Y = df.price
Y
```

```python
reg = LinearRegression()
reg.fit(X,Y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
# Predict price of a home with area = 3300 sqr ft
reg.predict([[3300]])
```

```python
# Y = m * X + b (m is coefficient and b is intercept)
3300*135.78767123 + 180616.43835616432
```

```python
# Predict price of a home with area = 5000 sqr ft
reg.predict([[5000]])
```

```python
model = LinearRegression()
model.fit(X, Y)
```

```python
y_pred = model.predict(X)
```

```python
plt.scatter(X, Y, color='blue')
plt.plot(X, y_pred, color='red', label='Regression Line')
plt.xlabel('Area')
plt.ylabel('Price')
plt.title('Linear Regression on homeprice Data')
plt.legend()
plt.show()
```

```python
X_future = np.array([[4500],[4698]])
y_future_pred = model.predict(X_future)
```

```python
plt.scatter(X, Y, color='blue', marker='+')
plt.plot(X, y_pred, color='red', label='Regression Line')
plt.scatter(X_future, y_future_pred, color='green', marker='o', label='Prediction')

plt.xlabel('Area')
plt.ylabel('Price')
plt.title('Linear Regression on HousePrices Data')
plt.legend()
plt.show()
```
