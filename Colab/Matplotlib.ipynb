```python
import numpy as np
import matplotlib.pyplot as plt

# Data
a = np.arange(5)
b = [2,4,6,8,10]
c = [5, 6, 7, 8, 9]

# Create plots
fig = plt.figure()
ax = plt.subplot()
ax.plot(a, b, 'k--', label='Frequency')
ax.plot(a, c, 'k:', label='Periods')

# Create a legend using the Matplotlib Axes.legend() method in Python
legend=ax.legend(loc='upper center',fontsize='xx-large')
legend.get_frame().set_facecolor('red')
# Plot Title
plt.title("Frequency of a Signal")

# Display
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data to plot
arr = np.array([10, 50, 34, 67, 21, 7, 59, 62, 45, 48, 10, 8, 41, 32, 66, 59, 18, 26, 51, 9])

# Plot a Histogram using pyplot.hist() method
# The bins parameter sets the bin i.e. an integer or sequence
plt.hist(arr, bins = [0, 20, 40, 60, 80])

# The labels for x-coordinate and y-coordinate
plt.xlabel("Score")
plt.ylabel("Player")

# Display the figure
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data to plot
# Score of two Teams
team1_Score = [25, 47, 34, 38, 27, 40, 42, 18]
team2_Score = [7, 22, 40, 29, 27, 10, 19, 31]

# Score Range of tailenders (lower order batsman) in Cricket
scoreRange = [5, 10, 15, 20, 25, 30, 35, 40]

# Plot a Scatter Plot using pyplot.scatter() method
plt.scatter(team1_Score, scoreRange, color='r')
plt.scatter(team2_Score, scoreRange, color='b')

# The labels for x-coordinate and y-coordinate
plt.xlabel("Team Score")
plt.ylabel("Score Range")

# Display the figure
plt.show()
```

```python
import matplotlib.pyplot as p
import matplotlib.image as mpimg
import pandas as pd

# Example data
df = pd.DataFrame({
    "Pname": ["Product A", "Product B", "Product C"],
    "Sales": [100, 150, 80]
})

x = df["Pname"]
y = df["Sales"]

# Create figure and axes
fig, ax = p.subplots(figsize=(8, 8))

# Bar plot
ax.bar(x, y, width=0.5)
ax.set_xlabel("Products")
ax.set_ylabel("Sales")
ax.tick_params(axis='x', labelrotation=90)

# Load image
image = mpimg.imread('Camera.jpeg')  # <-- Put your image file path here

# Create a new set of axes for the image above the main plot
new_ax = fig.add_axes([0.35, 0.80, 0.3, 0.15], anchor='N')  # [left, bottom, width, height]
new_ax.imshow(image)
new_ax.axis('off')  # Hide the axes frame

# Show plot
p.show()
```

```python
import matplotlib.pyplot as p
import pandas as pd

# Example data
df = pd.DataFrame({
    "Pname": ["Product A", "Product B", "Product C", "Product D"],
    "Sales": [150, 200, 180, 220]
})

x = df["Pname"]
y = df["Sales"]

# Create the bar plot
bars = p.bar(x, y, width=0.5)

# Add labels above each bar
for i, v in enumerate(y):
    p.text(i, v + 3, str(v), ha='center', va='bottom', fontsize="xx-small")

p.xlabel("Products")
p.ylabel("Sales")
p.xticks(rotation=90)
p.show()
```
