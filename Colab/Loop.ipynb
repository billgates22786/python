```python
# Write a program to check whether a number is prime or not
num=int(input("Enter a number: "))
k=0
if num==0 or num==1:
  print("Not a prime number")
else:
  for i in range(2,num):
    if num%i==0:
      k=k+1
if k==0:
  print(num,"is a prime number")
else:
  print(num,"is not a prime number")
```
