```python
k=open("Main.txt","r+")
# fileContent=input("Enter your contents: ")
# f.write(fileContent)
k.seek(0)
g=open("Vowels.txt","r+")
h=open("Consonants.txt","r+")
for i in k.read():
  if i in "aeiouAEIOU":
    g.write(i)
  else:
    h.write(i)
k.close()
g.seek(0)
h.seek(0)
print("Vowels:",g.read())
# print(g.read())
print("Consonants:",h.read())
# print(h.read())
g.close()
h.close()
```

```python
k=open("Main.txt","r+")
k.seek(0)
r=open("Reverse.txt","r+")
for i in k.read():
  r.write(i)
r.seek(0)
print(r.read())
```

```python
with open("random1.txt","w") as f:
  n=int(input("Enter the number of lines: "))
  for i in range(0,n):
    filecontent=input("Enter your contents: ")
    f.writelines(filecontent+"\n")
with open("random1.txt","r") as f:
  for i in f:
    print(i)
```
