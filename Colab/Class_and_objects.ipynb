```python
class Student:
  def __init__(self,name,rollno):
    self.name=name
    self.rollno=rollno
  def output(self):
    print("Student Name=",self.name)
    print("Student Rollno=",self.rollno)
  def __del__(self):
    print("Destructor Calling..")
```

```python
obj=Student("Sheery",1)
obj.output()
objj=Student("Gagan",2)
objj.output()
```
