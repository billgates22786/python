```python
# Function to print factorial
def factorial(x):
  f=1
  for i in range(x):
    f=f*(i+1)
    # dont use i=i+1 because it work on it own
  return f
k=int(input("Enter a number: "))
factorial(k)
```

```python
# Recursion function to print factorial
def factorial(gu):
  if gu==1:
    return 1
  else:
    return (gu*factorial(gu-1))
nu=int(input("Enter a number: "))
factorial(nu)
```
