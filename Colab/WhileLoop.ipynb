```python
# Write a program to check whether a number is prime or not using while loop
i=int(input("Enter a number: "))
m=0
if i==0 or i==1:
  print("Not a prime number ")
else:
  j=2
  while j<=i:
    if  i%j==0:
      m=m+1
    j=j+1
  if m==1:
    print("Prime number")
  else:
    print("Not a prime number")
```

```python
# Write a program to find the sum of the digits of a number accepted from the user
num=input("Enter any number:")
k=0
while num:
  k=k+int(num[0])
  num=num[1:]
print(f"Sum of the digits: {k}")
```

```python
# Write a program to find the product of the digits of a number accepted from user
num=input("Enter any number: ")
k=1
while num:
  k=k*int(num[-1])
  num=num[:-1]
print("Product of digits of number is: ",k)
```
