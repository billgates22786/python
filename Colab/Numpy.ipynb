**Convert a NumPy Array to 0 or 1 Array Based on Threshold Condition**

```python
import numpy as np
x1 = np.array(range(16))
x1 = x1.reshape([4,4])
print("x1=",x1)
y1 = np.where(x1<=4, 1, 0)
print(y1)
```

```python
import numpy as np
x1 = np.array(range(16))
x1 = x1.reshape([4,4])
print("x1=",x1)
x_mask = x1*3
y_mask = x1*-3
y1 = np.where(x1<=4, x_mask, y_mask)
print("y1=",y1)
y2 = np.where(np.logical_and(x1>=4, x1<=8), x_mask, y_mask)
print("y2=", y2)
```

**Unit-normalize a NumPy Array: A Practice Guide – NumPy Tutorial**

```python
import numpy as np
def unit(x, axis = 0):
    l2_norm = np.linalg.norm(x, keepdims=True, axis= axis) + 1e-12
    x_norm = x / l2_norm
    return x_norm
```

```python
x = np.random.random((10))
y = unit(x)
print(x)
print(y)
x1 = np.sum(y*y)
print(x1)
```

```python
x = np.random.random((10))
y = unit(x)
print(x)
print(y)
x1 = np.sum(y*y)
print(x1)
```

```python
import numpy as np
x = np.random.random((10, 10))
y = unit(x, axis = 1)
print(x)
print(y)
x1 = np.sum(np.square(y), axis = 1)
print(x1)
```

```python
# numpy.ndarray.byteswap()
import numpy as np
a = np.array([1, 256, 8755], dtype = np.int16)

print ('Our array is:')
print (a)
print ('\n')

print ('Representation of data in memory in hexadecimal form:')
print (map(hex,a))
print ('\n')
# byteswap() function swaps in place by passing True parameter

print ('Applying byteswap() function:')
print (a.byteswap(True))
print ('\n')

print ('In hexadecimal form:')
print (list (map(hex,a)) )
print ('\n')
# We can see the bytes being swapped
```
