```python
import csv
data=[
    ['Kartikay',1,19]
    ]
with open('Students.csv',"w") as f:
    writer=csv.writer(f)
    writer.writerows(data)

print("CSV without header created.")
```

```python
import csv
with open('Students.csv',"r") as f:
    data = list(csv.reader(f))

with open('students_with_header.csv',"w") as of:
    writer = csv.writer(of)
    writer.writerow(['name', 'rollno', 'age'])
    writer.writerows(data)

print("CSV with header created as 'students_with_header.csv'.")
```
