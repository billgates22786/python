```python
a=int(input("Enter first number: "))
b=int(input("Enter second number:"))
print("****Choice****\n1.Addition\n2.Subtraction\n3.Multiplication\n4.Division")
c=int(input("Enter your choice: "))
match c:
    case 1:
        print(f"Addition of the two numbers : {a+b}")
    case 2:
        print(f"Subtraction of the two numbers: {a-b}")
    case 3:
        print(f"Multiplication of the two numbers: {a*b}")
    case 4:
        if b!=0:
            print(f"Division of the two numbers: {a/b}")
        else:
            print("Invalid Value")
        
```

```python
num=int(input("Enter a number: "))
c=num%2
match c:
    case 0:
      print("Even Number")
    case _: 
      print("Odd Number")
```

```python
cha=input("Enter a character: ")
match cha:
    case "a"|"A"|"e"|"E"|"i"|"I"|"o"|"O"|"u"|"U":
        print("Vowel")
    case _:
        print("Consonant")
```
